```python
import pandas as pd
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Fungsi untuk membaca daftar kata dari file .txt
def load_words(file_path):
    with open(file_path, 'r') as file:
        words = file.read().splitlines()
    return words

# Fungsi untuk membaca kata slang dari file .txt
def load_slang_words(file_path):
    slang_dict = {}
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split(',')
            if len(parts) == 2:
                slang, proper = parts
                slang_dict[slang] = proper
    return slang_dict

# Fungsi untuk menghapus karakter khusus dari tweet
def remove_tweet_special(text): 
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    text = text.encode('ascii','replace').decode('ascii')
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    text = re.sub(r'^RT[\s]+', '', text)
    text = re.sub(r',','',text)
    return text.replace("http://", " ").replace("https://", " ")

# Fungsi untuk menghapus angka dari teks
def remove_number(text):
    return re.sub(r"\d+", "", text)

# Fungsi untuk menghapus tanda baca dari teks
def remove_punctuation(text): 
    return text.translate(str.maketrans("", "", string.punctuation))

# Fungsi untuk menghapus whitespace di awal dan akhir teks
def remove_whitespace_LT(text):
    return text.strip()

# Fungsi untuk menghapus whitespace berlebih
def remove_whitespace_multiple(text):
    return re.sub('\s+', ' ', text)

# Fungsi untuk menghapus karakter tunggal
def remove_single_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)

# Fungsi untuk menghapus stopword
def remove_stopword(text):
    stop_words = set(stopwords.words('stopwords_indonesia'))
    word_tokens = word_tokenize(text)
    filtered_sentence = [w for w in word_tokens if not w in stop_words]
    return ' '.join(filtered_sentence)

# Fungsi untuk mengganti kata slang dengan kata yang benar
def slangword(text, slang_dict):
    word_tokens = word_tokenize(text)
    replace_sentence = [slang_dict.get(w.lower(), w) for w in word_tokens]
    return ' '.join(replace_sentence)

# Fungsi untuk stemming
def stemming(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    return stemmer.stem(text)

# Fungsi untuk preprocessing teks
print("jalan")
```

```python
def preprocess_text(text, slang_dict):
    text = remove_tweet_special(text)
    text = remove_number(text)
    text = remove_punctuation(text)
    text = remove_whitespace_LT(text)
    text = remove_whitespace_multiple(text)
    text = remove_single_char(text)
    text = remove_stopword(text)
    text = slangword(text, slang_dict)
    text = stemming(text)
    return text

print('jalan')
```

```python
# Fungsi untuk melabeli sentimen teks
def label_sentiment(text, positive_words, negative_words):
    words = word_tokenize(text)
    positive_score = sum(1 for word in words if word in positive_words)
    negative_score = sum(1 for word in words if word in negative_words)
    
    if positive_score > negative_score:
        return 'positif'
    elif positive_score < negative_score:
        return 'negatif'
    else:
        return 'netral'

# Path ke file yang berisi daftar kata slang dan padanannya
slang_file_path = 'slangword.txt'
positive_words_path = 'positive.txt'
negative_words_path = 'negative.txt'

# Memuat daftar kata slang, positif, dan negatif
slang_dict = load_slang_words(slang_file_path)
positive_words = load_words(positive_words_path)
negative_words = load_words(negative_words_path)

print("jalan")
```

```python
# Preprocessing dan pelabelan sentimen pada setiap teks dalam dataset
df['processed_text'] = df['Komentar'].apply(lambda x: preprocess_text(x, slang_dict))
df['sentiment'] = df['processed_text'].apply(lambda x: label_sentiment(x, positive_words, negative_words))
```
